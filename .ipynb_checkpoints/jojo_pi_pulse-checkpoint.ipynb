```python
%matplotlib inline
#from python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import distance
from scipy.optimize import curve_fit
# import cma

from qiskit import IBMQ
import qiskit.pulse as pulse
import qiskit.pulse.pulse_lib as pulse_lib
from qiskit.compiler import assemble
import qiskit.qobj.utils as qqu

%load_ext autoreload
%autoreload 2
```

```python
IBMQ.load_account()
provider = IBMQ.get_provider(hub='ibm-q-ornl', group='ornl', project='phy141')
backend_name = 'ibmq_johannesburg'
backend      = provider.get_backend(backend_name)
config       = backend.configuration()
defaults     = backend.defaults()
```

```python
GHz = 1.0e9
MHz = 1.0e6
us = 1.0e-6
ns = 1.0e-9
dt = config.dt 

qubit  = 0 
proper_frequency_Hz = defaults.qubit_freq_est[qubit]
scale_factor = 1e-14

### collect the necessary channels
drive_channel = pulse.DriveChannel(qubit)
meas_channel = pulse.MeasureChannel(qubit)
acq_channel = pulse.AcquireChannel(qubit)

def get_closest_multiple_of_16(num):
    return int(num + 8) - (int(num +8)%16)

#Drive pulse parameters (us = microseconds)
drive_sigma_us = 0.0064
drive_samples_us = drive_sigma_us*8

drive_sigma = get_closest_multiple_of_16(drive_sigma_us * us / dt)
drive_samples = get_closest_multiple_of_16(drive_samples_us * us / dt)

print(drive_sigma)
print(drive_samples)
```

```python
### calibrating the Pi pulse
num_rabi_points = 71
drive_amp_min = 0
drive_amp_max = 1.
drive_amps = np.linspace(drive_amp_min, drive_amp_max, num_rabi_points)

#Find out which group of qubits need to be acquire with this qubit
meas_map_idx = None
for i , measure_group in enumerate(config.meas_map):
    if qubit in measure_group:
        meas_map_idx = i
        break
assert meas_map_idx is not None, f"couldn't find qubit {qubit} in the meas_map!"

#now we can define our measurement pulse

inst_sche_map = defaults.instruction_schedule_map
measure = inst_sche_map.get('measure', qubits = config.meas_map[meas_map_idx])
```

```python
rabi_schedules = []
for drive_amp in drive_amps:
    rabi_pulse = pulse_lib.gaussian(duration = drive_samples, 
                                     amp = drive_amp,
                                     sigma = drive_sigma,
                                     name = f"Rabi drive amplitude = {drive_amp}")
    this_schedule = pulse.Schedule(name=f"Rabi drive amplitude = {drive_amp}")
    this_schedule += pulse.Play(rabi_pulse, drive_channel)
    this_schedule += measure << this_schedule.duration
    rabi_schedules.append(this_schedule)
```

```python
num_shots_per_point = 1024

rabi_experiment_program = assemble(rabi_schedules,
                                   backend=backend,
                                   meas_level=1,
                                   meas_return='avg',
                                   shots=num_shots_per_point,
                                   schedule_los=[{drive_channel: proper_frequency_Hz}]* num_rabi_points)
```

```python
from qiskit.tools.monitor import job_monitor
job = backend.run(rabi_experiment_program)
job_monitor(job)
```

```python
rabi_results = job.result(timeout=1200)
```

```python
def get_amplitude(vec):
    i_signal = np.imag(vec)
    r_signal = np.real(vec)
    
    r_max = np.max(r_signal)
    r_min = np.min(r_signal)
    i_max = np.max(i_signal)
    i_min = np.min(i_signal)

    r_m = 0.5*(r_max + r_min)
    i_m = 0.5*(i_max + i_min)

    mvec = [r_m, i_m]
    src_mat = np.vstack((r_signal - mvec[0], i_signal - mvec[1])).T
    ( _, _, v_mat) = np.linalg.svd(src_mat)
    # A = U D V^H
    # A V = U D = E 
    # the first column of E has the strongly correlated information
    dvec = v_mat[0, 0:2] # here we keep the first column of E 
    if dvec.dot(mvec) < 0:
        dvec = -dvec
    return src_mat.dot(dvec)
```

```python
 from scipy.optimize import curve_fit 
 
 def fit_function(tt, yy):
    
        tt = np.array(tt)
        yy = np.array(yy)
        ff = np.fft.fftfreq(len(tt), (tt[1]-tt[0]))   # assume uniform spacing
        Fyy = abs(np.fft.fft(yy))
        guess_freq = abs(ff[np.argmax(Fyy[1:])+1])   # excluding the zero frequency "peak", which is related to offset
        guess_amp = np.std(yy) * 2.**0.5
        guess_offset = np.mean(yy)
        guess = np.array([guess_amp, 2.*np.pi*guess_freq, 0., guess_offset])

        def cosfunc(t, A, w, p, c):  return A * np.cos(w*t + p) + c
        popt, pcov = curve_fit(cosfunc, tt, yy, p0=guess)
        A, w, p, c = popt
        f = w/(2.*np.pi)
        fitfunc = lambda t: A * np.cos(w*t + p) + c
        return {"amp": A, "omega": w, "phase": p, "offset": c, "freq": f, "period": 1./f, "fitfunc": fitfunc, "maxcov": np.max(pcov), "rawres": (guess,popt,pcov)}

```

```python
fit_results = fit_function(drive_amps, rabi_values)
```

```python
y_fit  = fit_results['fitfunc']
drive_period = fit_results['period']
freq = fit_results['omega']
phase = fit_results['phase']/freq
plt.scatter(drive_amps, rabi_values, color='black')
plt.plot(drive_amps, y_fit(drive_amps))
plt.axvline( -phase, color = 'red', linestyle = '--')
plt.axvline( drive_period/2-phase, color = 'red', linestyle = '--')
plt.annotate("", xy=(-phase, 0), xytext=(drive_period/2 - phase,0), arrowprops=dict(arrowstyle="<->", color='red'))
plt.annotate("$\pi$", xy=(-phase-0.05, 0.1), color='red')
plt.xlabel("Drive amp [a.u.]", fontsize=15)
plt.ylabel("Measured signal [a.u.]", fontsize=15)
plt.show()
```

```python
#creating pulses
circ_inst_map = defaults.instruction_schedule_map
measure = circ_inst_map.get('measure', qubits=config.meas_map[0])
```

```python
def get_amplitude(vec):
    i_signal = np.imag(vec)
    r_signal = np.real(vec)

    mvec = [np.mean(r_signal), np.mean(i_signal)]

    src_mat = np.vstack((r_signal - mvec[0], i_signal - mvec[1])).T
    (_, _, v_mat) = np.linalg.svd(src_mat)

    dvec = v_mat[0, 0:2]

    if dvec.dot(mvec) < 0:
        dvec = -dvec

    return src_mat.dot(dvec)
```

```python
#creating a measurement pulse
m0         = 0.05
m_samples  = 1200
m_sigma    = 4
m_risefall = 25
m_pulse   = pulse_lib.gaussian_square(duration = m_samples,
                                           amp      = m0,
                                           sigma    = m_sigma,
                                           risefall = m_risefall,
                                           name     = 'measure')
#create an acquire pulse
acq_cmd = pulse.Acquire(duration=m_samples)

meas_map_idx = None
for i, measure_group in enumerate(config.meas_map):
    if qubit in measure_group:
        meas_map_idx = i
        break
        
assert meas_map_idx is not None, f"Couldn't find qubit {qb} in the meas_map!"
        
#define channels
m_ch = pulse.channels.MeasureChannel(qubit)

meas = m_pulse(m_ch)
meas += acq_cmd([pulse.AcquireChannel(i) for i in config.meas_map[meas_map_idx]],
                    [pulse.MemorySlot(i) for i in config.meas_map[meas_map_idx]])        
```

```python
#running the experiment
qubit = 0
exps = 64
shots = 512

drive_amps = np.linspace(0,0.1,exps)
drive_samples = 128
drive_sigma = 16
scale_factor = 1e-10
```

```python
schedules = []
for ii, drive_amp in enumerate(drive_amps):
    rabi_pulse = pulse_lib.gaussian(duration=drive_samples, 
                                    amp=drive_amp,
                                    sigma=drive_sigma,
                                    name='rabi_pulse_%d' % ii)
    schedule = pulse.Schedule(name='Rabi Experiment at drive amp = %s' % drive_amp)
    schedule |= rabi_pulse(pulse.DriveChannel(qubit))
    schedule |= meas << schedule.duration
    
    schedules.append(schedule)
```

```python
schedules[-1].draw(channels=[pulse.DriveChannel(qubit), pulse.MeasureChannel(qubit)],
                   label=True, plot_range=(0, 4096))
```

```python
rabi_qobj = assemble(schedules, backend, 
                     meas_level=qqu.MeasLevel.KERNELED, 
                     meas_return=qqu.MeasReturnType.AVERAGE, 
                     shots=shots)
```

```python
job = backend.run(rabi_qobj)
```
